```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
```

```python
data_path_onefeature = "student_scores.csv"
data_path_multifeature = "Data.csv"
```

```python
def evaluation_testsize(X, Y):
    list_score = []
    for test in range(1, 6):
        x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=(test/10), random_state=0)
        regressor = LinearRegression()
        regressor.fit(x_train, y_train)
        y_pred = regressor.predict(x_test)
        R2_SCORE = metrics.r2_score(y_test, y_pred)
        list_score.append(R2_SCORE)
    return list_score

print(evaluation_testsize(X_multifeature, Y_multifeature))


plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.plot(evaluation_testsize(X_onefeature, Y_onefeature), 'o--')
plt.title("evaluetion test size onefeature")
plt.xlabel("test size")
plt.ylabel("Scores")
plt.xticks(range(5), ['90-10', '80-20', '70-30', '60-40', '50-50'])
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(evaluation_testsize(X_multifeature, Y_multifeature), 'o--')
plt.title("evaluetion test size multifeature")
plt.xlabel("test size")
plt.ylabel("Scores")
plt.xticks(range(5), ['90-10', '80-20', '70-30', '60-40', '50-50'])
plt.grid()
```

```python
def test_newfeature(df):
    evaluation_list = []
    for power in range(2, 25):
        Hours = df['TeamWork']**power
        df.insert(3, str(power), Hours)
        
        X = df[[df.columns[0], df.columns[1], df.columns[2], df.columns[3]]]
        Y = df[df.columns[4]]
        
        x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= .4, random_state=0)
        
        regressor = LinearRegression()
        regressor.fit(x_train, y_train)
        
        y_pred = regressor.predict(x_test)
        
        R2_SCORE = metrics.r2_score(y_test, y_pred)
        if R2_SCORE >= .99:
            print("overfit")
        else:
            evaluation_list.append(R2_SCORE)
            df.drop([str(power)], axis = 1, inplace = True)
        
    return evaluation_list


def display_plot(function):
    plt.figure(figsize=(10, 8))
    plt.plot(function , 'o--')
    plt.title("evaluation degree faeture TeamWork for find best")
    plt.xlabel('degree')
    plt.ylabel("R2_Score")
    plt.xticks(range(len(function)))
    plt.grid()
    return plt.show()
    
display_plot(test_newfeature(df_multifeature))
print(max(test_newfeature(df_multifeature)))
print(len(test_newfeature(df_multifeature)))
```
