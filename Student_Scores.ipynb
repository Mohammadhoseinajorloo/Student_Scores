```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data_path_onefeature = "student_scores.csv"
data_path_multifeature = "Data.csv"
```

```python
data_onefeature = pd.read_csv (data_path_onefeature)
```

```python
df_onefeature = pd.DataFrame(data_onefeature)
```

```python
df_onefeature.dropna(inplace=True)
```

```python
plt.scatter(df_onefeature.Hours, df_onefeature.Scores)
plt.title("plot Hours per Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.grid()
plt.savefig("Hours v Scores")
plt.show()
```

```python
X_onefeature = df_onefeature['Hours'].values.reshape (-1, 1)
Y_onefeature = df_onefeature["Scores"].values.reshape (-1, 1)
```

```python
x_trin_onefeature, x_test_onefeature, y_trin_onefeature, y_test_onefeature = train_test_split(X_onefeature, Y_onefeature, test_size=.2, random_state=0)
```

```python
regressor_onefeature = LinearRegression()
```

```python
regressor_onefeature.fit(x_trin_onefeature, y_trin_onefeature)
```

```python
y_pred_onefeature = regressor_onefeature.predict(x_test_onefeature)
```

```python
intercept_onefeature = regressor_onefeature.intercept_
coef_onefeature = regressor_onefeature.coef_
```

```python
intercept_onefeature
```

```python
coef_onefeature
```

```python
plt.scatter(x_test_onefeature, y_test_onefeature, color='gray')
plt.plot(x_test_onefeature, y_pred_onefeature, color='red')
plt.title("plot actual data and prediction data")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.grid()
plt.savefig("actual data v prediction data(plot)")
plt.show()
```

```python
MAE_onefeature = metrics.mean_absolute_error(y_test_onefeature, y_pred_onefeature)
MSE_onefeature = metrics.mean_squared_error(y_test_onefeature, y_pred_onefeature)
RMSE_onefeature = np.sqrt(MSE_onefeature)
R2_SCORE_onefeature = metrics.r2_score(y_test_onefeature, y_pred_onefeature)
```

```python
print(MAE_onefeature)
print(MSE_onefeature)
print(RMSE_onefeature)
print(R2_SCORE_onefeature)
```
